```python
#  Cuántas veces aparecía el día de la semana en cada categoría de textos corpus.
from nltk.corpus import brown

cond_frq_dist = nltk.ConditionalFreqDist(
    (genre, word.lower())
    for genre in brown.categories()
    for word in brown.words(categories=genre)
)
days_of_week = ["monday", "tuesday", "wednesday", "thursday", "friday", "saturday", "sunday"]
genres = brown.categories()
cond_frq_dist.tabulate(conditions=genres, samples=days_of_week)
```

```python
# Graficar lo de arriba
cond_frq_dist.plot(conditions=genres, samples=days_of_week)
```
